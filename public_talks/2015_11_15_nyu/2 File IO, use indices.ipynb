# File I/O

## Get all paths to TLG or PHI5

```python
from cltk.corpus.utils.formatter import assemble_phi5_author_filepaths
```

```python
# For author files
phi_fps = assemble_phi5_author_filepaths()
```

```python
print(phi_fps[:10])
```

```python
# For individual files
from cltk.corpus.utils.formatter import assemble_phi5_works_filepaths
phi_author_fps = assemble_phi5_works_filepaths()
print(phi_author_fps[:10])
```

## Example: Open and read files

```python
# Open and print to screen the first 100 characters of the first 5 individual texts from the PHI5
count = 0
for fp in phi_author_fps:
    count += 1
    if count < 5:
        with open(fp) as file_open:
            file_read = file_open.read()
            print(file_read[:100])
```

Getting all texts is useful only if you want to operate on entirety of a corpus.

```python
# Get all historians from Halicarnassus
id_author_dict = get_id_author()  # global index

halicarnissi = select_authors_by_geo('Halicarnassus')
for _id in halicarnissi:
    print(_id, id_author_dict[_id])
```

```python
# Build filepaths and open Halicarnissi
import os

author_fps_base = os.path.expanduser('~/cltk_data/greek/text/tlg/plaintext/')
for _id in halicarnissi:
    
    fp = os.path.join(author_fps_base, 'TLG' + _id + '.txt')
    print('Opening file:', fp)

    with open(fp) as fo:
        fr = fo.read()
    print(id_author_dict[_id])
    print(fr[:100])
    print()
    print()
```
