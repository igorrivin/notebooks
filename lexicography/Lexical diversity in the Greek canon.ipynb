```python
from nltk.tokenize.punkt import PunktWordTokenizer
import os
import re
import sys
```

```python
def join_hyphens(str_text):
    return str_text.replace('-\n', '')

def tokenize(str_text):
    tokens = PunktWordTokenizer().tokenize(str_text)
    tokens = [tokens.pop() for x in tokens if x not in (',', ';', '.', "'", '"')]
    return tokens

def lexical_diversity(tokenized_words):
    """This is the main calculation. Unique words divided by total words.
    Higher number means greater diversity.
    """
    return len(set(tokenized_words)) / len(tokenized_words)

def process_text(file_path):
    with open(file_path) as f:
        r = f.read()
        r = join_hyphens(r)
        return tokenize(r)
```

```python
def build_tlg_index():
    """Returns dict of {file: author_name}
    Should return 1823 files.
    """
    index_path_rel = '~/cltk_data/originals/tlg/AUTHTAB.DIR'
    index_path = os.path.expanduser(index_path_rel)
    with open(index_path, 'rb') as f:
        r = f.read()
        index_all = r.decode('latin-1').split('\xff')[1:-6]
        index = [x for x in index_all if x]
        file_author = {}
        for x in index:
            # file name
            pattern_file = re.compile('TLG[\d].{4}')
            m = pattern_file.match(x)
            file_name = m.group()[:-1] + '.TXT'

            # author name
            author_name = pattern_file.split(x)[-1]
            pattern_author = re.compile('[\d]|&')
            author_name = pattern_author.sub('', author_name)
            pattern_comma = re.compile('\x80')
            author_name = pattern_comma.sub(', ', author_name)
            file_author[file_name] = author_name

    return file_author
```

```python
def build_scores(min_tokens=0):
    """Returns dict of {file: score}. 
    Should return 1823 if min_tokens=0.
    """
    tlg_dir_rel = '~/cltk_data/greek/text/tlg/plaintext/'
    tlg_dir = os.path.expanduser(tlg_dir_rel)
    tlg_files = os.listdir(tlg_dir)
    #tlg_files = [tlg_files.pop() for x in tlg_files if x.startswith('TLG')]
    tlg_files = [x for x in tlg_files if x.startswith('TLG')]
    
    file_score = {}
    for tlg_file in tlg_files:
        tlg_file_path = os.path.join(tlg_dir, tlg_file)
        tokens = process_text(tlg_file_path)
        
        if len(tokens) < min_tokens:
            continue
        else:
            score = lexical_diversity(tokens)
            file_score[tlg_file] = score
    
    return file_score
```

```python
tlg_index = build_tlg_index()
tlg_scores = build_scores()
```

```python
def join_dicts(file_author, file_score):
    """now do join on two dicts"""
    author_score = {}
    for x in file_author:
        if x in file_score:
            name = file_author[x]
            score = file_score[x]
            author_score[name] = score
    return author_score
```

```python
author_score = join_dicts(tlg_index, tlg_scores)
#print(author_score)
```

```python
import pandas as pd

#display enough rows for all authors
pd.set_option('display.max_rows', 10000)

# generate the Pandas DataFrame
df = pd.DataFrame(author_score, index=['Lexical diversity'])

print(df.T.sort('Lexical diversity', ascending=0))

# All TLG texts below.
# This arguably returns too many frangmentary texts without enough text to make a fair sample of.
# See next cell for the table, only excluding texts with less than 1000 words.
```

```python
tlg_scores = build_scores(min_tokens=1000)
```

```python
author_score = join_dicts(tlg_index, tlg_scores)

print('Texts with more than 1000 words:', len(tlg_scores))

#display enough rows for all authors
pd.set_option('display.max_rows', 10000)

# generate the Pandas DataFrame
df = pd.DataFrame(author_score, index=['Lexical diversity'])

print(df.T.sort('Lexical diversity', ascending=0))
```
