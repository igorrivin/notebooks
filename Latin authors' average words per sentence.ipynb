# Tokenize and count

This first step tokenizes words and sentences, then returns the ratio of average words per sentence for each author in the Packard Humanities Institute's 5th disc ([corpus also available online](http://latin.packhum.org/)). The tokenization script ran about 15 minutes on my machine ([download here](https://gist.github.com/kylepjohnson/88ce358c95f8cb668e54/download)). It generates a file called called `phi5_auth_sent_data.txt` ([download](https://gist.github.com/kylepjohnson/794636f05795d81ef04a/download)).

``` python
"""Complete class and script for computing average sentence length for PHI5
authors.
"""

import ast
from cltk.tokenizers.sentence_tokenizer import train_and_tokenize_latin
from nltk.tokenize import RegexpTokenizer
import os


# PHI5 parameters
PHI5_DIR = os.path.join('~/cltk_data', 'compiled', 'phi5')
PHI5_INDEX_REL = os.path.join(PHI5_DIR, 'index_author_works.txt')

# Local write parameters
WRITE_FILE_NAME = 'phi5_auth_sent_data.txt'
WRITE_DIR_LOCATION = '~/Downloads'


class AvgSents(object):
    """Contains methods for computing average sentence length."""

    def open_file(self, abs_path):
        """Open any file with an absolute path."""
        try:
            with open(abs_path) as file:
                read_file = file.read()
            return read_file
        except:
            pass

    def phi5_author_path(self, author, index_dict):
        """Build absolute path to PHI5 author–file."""
        phi5_file = index_dict[author]['phi5_file']#?
        phi5_file_txt = os.path.join(phi5_file + '.txt')
        phi5_file_rel = os.path.join(PHI5_DIR, phi5_file_txt)
        phi5_file_abs = os.path.expanduser(phi5_file_rel)
        return phi5_file_abs

    def count_tokenize_sents(self, author_read):
        """Tokenize and count sentences in a string."""
        try:
            sents = train_and_tokenize_latin(author_read)
        except:
            pass
        return len(sents)

    def write_dict(self, auth_s_w_dict):
        """Write author sentence–word data into dict in a file."""
        write_rel_path = os.path.join(WRITE_DIR_LOCATION, WRITE_FILE_NAME)
        write_abs_path = os.path.expanduser(write_rel_path)
        try:
            with open(write_abs_path, 'w') as file:
                try:
                    file.write(str(auth_s_w_dict))
                except:
                    pass
        except:
            pass

    def count_tokenize_words(self, author_read):
        """Tokenize and count words in a string."""
        tokenizer = RegexpTokenizer('\s+', gaps=True)
        try:
            words = tokenizer.tokenize(author_read)
            return len(words)
        except:
            pass


def main():
    """Main function"""

    avg = AvgSents()

    phi5_index_abs = os.path.expanduser(PHI5_INDEX_REL)
    print(phi5_index_abs)
    index_read = avg.open_file(phi5_index_abs)
    index_dict = ast.literal_eval(index_read)

    auth_s_w_dict = {}
    for author in index_dict:
        phi5_file_abs = avg.phi5_author_path(author, index_dict)
        author_read = avg.open_file(phi5_file_abs)
        try:
            sent_count = avg.count_tokenize_sents(author_read)
            print(sent_count)
        except:
            sent_count = 0
        try:
            word_count = avg.count_tokenize_words(author_read)
        except:
            word_count = 0
        try:
            ratio = word_count / sent_count
        except:
            ratio = 0
        counts = {'sent_count': sent_count,
                  'word_count': word_count,
                  'ratio': ratio}
        auth_s_w_dict[author] = counts

    avg.write_dict(auth_s_w_dict)

if __name__ == "__main__":
    main()

```

# Views

Next, I offer some simple views of the file `phi5_auth_sent_data.txt` using the data analysis library Pandas.

````python
import ast
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd

#display all PHI5 authors
pd.set_option('display.max_rows', 10000)
````

````python
file_name = 'phi5_auth_sent_data.txt'
dir_location = '~/Downloads'
rel_path = os.path.join(dir_location, file_name)
abs_path = os.path.expanduser(rel_path)
with open(abs_path) as f:
    r = f.read()
d = ast.literal_eval(r)
````

````python
# generate the Pandas DataFrame
df = pd.DataFrame(d)
````

````python
df.T
````

## Sorted by average words per sentence

The ratio of the first three authors listed below (Manius Manilius, Mimi Poetarum Incertorum, and Decimus Laberius) is suspiciously high, possibly signaling an issue in the markup of these fragmentary texts. Beyond these aberrations the data appears at first blush to be plausible.

````python
# Ordered descending by words per sentence ratio
df.T.sort(['ratio', 'sent_count', 'word_count'], ascending=[0, 0, 0])
````
