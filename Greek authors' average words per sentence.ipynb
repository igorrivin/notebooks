# Tokenize and count

This first step tokenizes words and sentences, then returns the ratio of average words per sentence for each author in the Thesaurus Linguae Grecae. The following code runs too long for an IPython notebook (over an hour to run on my machine) though it is nevertheless readable ([download code here](https://gist.github.com/kylepjohnson/32aa32ba5d3d288b7a25/download)). It generates a file called called `tlg_auth_sent_data.txt` ([download](https://gist.github.com/kylepjohnson/d317ccfa82a743f43b55/download)).

``` python
"""Complete class and script for computing average sentence length for TLG
authors.
"""

import ast
from cltk.tokenizers.sentence_tokenizer import train_and_tokenize_greek
from nltk.tokenize import RegexpTokenizer
import os


# TLG parameters
TLG_DIR = os.path.join('~/cltk_data', 'compiled', 'tlg')
TLG_INDEX_REL = os.path.join(TLG_DIR, 'index_author_works.txt')

# Local write parameters
WRITE_FILE_NAME = 'tlg_auth_sent_data.txt'
WRITE_DIR_LOCATION = '~/Downloads'


class AvgSents(object):
    """Contains methods for computing average sentence length."""

    def open_file(self, abs_path):
        """Open any file with an absolute path."""
        try:
            with open(abs_path) as file:
                read_file = file.read()
            return read_file
        except:
            pass

    def tlg_author_path(self, author, index_dict):
        """Build absolute path to TLG author–file."""
        tlg_file = index_dict[author]['tlg_file']
        tlg_file_txt = os.path.join(tlg_file + '.txt')
        tlg_file_rel = os.path.join(TLG_DIR, tlg_file_txt)
        tlg_file_abs = os.path.expanduser(tlg_file_rel)
        return tlg_file_abs

    def count_tokenize_sents(self, author_read):
        """Tokenize and count sentences in a string."""
        try:
            sents = train_and_tokenize_greek(author_read)
        except:
            pass
        return len(sents)

    def write_dict(self, auth_s_w_dict):
        """Write author sentence–word data into dict in a file."""
        write_rel_path = os.path.join(WRITE_DIR_LOCATION, WRITE_FILE_NAME)
        write_abs_path = os.path.expanduser(write_rel_path)
        try:
            with open(write_abs_path, 'w') as file:
                try:
                    file.write(str(auth_s_w_dict))
                except:
                    pass
        except:
            pass

    def count_tokenize_words(self, author_read):
        """Tokenize and count words in a string."""
        tokenizer = RegexpTokenizer('\s+', gaps=True)
        try:
            words = tokenizer.tokenize(author_read)
            return len(words)
        except:
            pass


def main():
    """Main function"""

    avg = AvgSents()

    tlg_index_abs = os.path.expanduser(TLG_INDEX_REL)
    index_read = avg.open_file(tlg_index_abs)
    index_dict = ast.literal_eval(index_read)

    auth_s_w_dict = {}
    for author in index_dict:
        tlg_file_abs = avg.tlg_author_path(author, index_dict)
        author_read = avg.open_file(tlg_file_abs)
        try:
            sent_count = avg.count_tokenize_sents(author_read)
            print(sent_count)
        except:
            sent_count = 0
        try:
            word_count = avg.count_tokenize_words(author_read)
        except:
            word_count = 0
        try:
            ratio = word_count / sent_count
        except:
            ratio = 0
        counts = {'sent_count': sent_count,
                  'word_count': word_count,
                  'ratio': ratio}
        auth_s_w_dict[author] = counts

    avg.write_dict(auth_s_w_dict)

if __name__ == "__main__":
    main()
```

# Views

Next, I offer some simple views of the file `tlg_auth_sent_data.txt` using the data analysis library Pandas. These tables are alphabetical by author name, global average words per sentence (highest to lowest) and averages within genres (lowest to highest), average words per sentence, total sentences (highest to lowest), and total words (highest to lowest).

````python
import ast
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd

#display all TLG authors
pd.set_option('display.max_rows', 10000)
````

````python
file_name = 'tlg_auth_sent_data.txt'
dir_location = '~/Downloads'
rel_path = os.path.join(dir_location, file_name)
abs_path = os.path.expanduser(rel_path)
with open(abs_path) as f:
    r = f.read()
d = ast.literal_eval(r)
````

````python
# generate the Pandas DataFrame
df = pd.DataFrame(d)
````

## Sorted by author name

````python
df.T
````

## Sorted by average words per sentence

From the following table, it is clear that the sentences of the first dozen or so authors are not being counted correctly, most likely due to editing conventions of fragmentary texts (e.g., Agathon, Philodemus). Beyond these aberrations the data appears at first blush to be plausible.

````python
# Ordered ascending by words per sentence ratio
df.T.sort(['ratio', 'sent_count', 'word_count'], ascending=[0, 0, 0])
````

### Historians by avg. words per sentence

With only the TLG's labels, we can sort by genre. Consider, for example, authors labeled as historians ('Hist.') by the TLG.

````python
authors = list(df.columns.values)

# generate a new, historian-only dict
# there's probably a more graceful way to do this with Pandas and the
# previous dict
dict_gen = {}
for author in authors:
    if ' Hist.' in author:
        dict_gen[author] = d[author]
df_hist = pd.DataFrame(dict_gen)

df_hist.T.sort(['ratio', 'sent_count', 'word_count'], ascending=[1, 0, 0])
````
